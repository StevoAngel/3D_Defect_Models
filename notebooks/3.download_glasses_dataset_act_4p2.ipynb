```python
import os
import shutil
import kagglehub
import pandas as pd
```

```python
# Download latest version
path = kagglehub.dataset_download("jeffheaton/glasses-or-no-glasses")

print("Path to dataset files:", path)
```

```python
train = pd.read_csv(f"{path}/train.csv")
train.set_index('id', inplace=True)
train.tail()
```

```python
# Devide and move images to respective folders according to glasses or no glasses:

wGlasses = "../data/raw/wGlasses" # Images with glasses
woGlasses = "../data/raw/woGlasses" # Images without glasses

os.makedirs(wGlasses, exist_ok=True) # Create directory if it does not exist
os.makedirs(woGlasses, exist_ok=True) # Create directory if it does not exist

_path = path.replace("\\", "/") # Ensure path uses forward slashes

folder = f"{_path}/faces-spring-2020/" # The folder where the images are located

for i in range(len(train)):
    oldpath = f"{folder}face-{i}.png" # Original image path
    if train.iloc[i]['glasses'] == 0: # No glasses
        newpath = os.path.join(woGlasses, f"face{i}.png") # New image path
    elif train.iloc[i]['glasses'] == 1: # Has glasses
        newpath = os.path.join(wGlasses, f"face{i}.png") # New image path

    if os.path.exists(oldpath):
        shutil.move(oldpath, newpath) # Move the file to the new path
    else:
        print(f"File not found: {oldpath}")
```
